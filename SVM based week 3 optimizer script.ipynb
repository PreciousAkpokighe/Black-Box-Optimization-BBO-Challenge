```python
import numpy as np

# =========================
# WEEK 1 INPUTS
# =========================
week1_inputs = [
    np.array([0.211325, 0.788675]),
    np.array([0.723607, 0.276393]),
    np.array([0.166667, 0.500000, 0.833333]),
    np.array([0.125000, 0.375000, 0.625000, 0.875000]),
    np.array([0.875000, 0.625000, 0.375000, 0.125000]),
    np.array([0.150000, 0.350000, 0.550000, 0.750000, 0.950000]),
    np.array([0.900000, 0.100000, 0.700000, 0.300000, 0.500000, 0.800000]),
    np.array([0.111111, 0.222222, 0.333333, 0.444444, 0.555555, 0.666667, 0.777778, 0.888889])
]

# =========================
# WEEK 2 INPUTS
# =========================
week2_inputs = [
    np.array([0.654299, 0.353479]),
    np.array([0.754299, 0.253479]),
    np.array([0.304299, 0.553479, 0.709691]),
    np.array([0.804299, 0.603479, 0.409691, 0.205016]),
    np.array([0.854299, 0.653479, 0.359691, 0.155016]),
    np.array([0.904299, 0.703479, 0.509691, 0.305016, 0.103275]),
    np.array([0.884299, 0.123479, 0.729691, 0.285016, 0.523275, 0.787492]),
    np.array([0.104299, 0.243479, 0.319691, 0.465016, 0.573275, 0.647492, 0.794889, 0.860861])
]

# =========================
# WEEK 1 OUTPUTS
# =========================
week1_outputs = np.array([
    1.1327056288856873e-125,
    0.5675786892822564,
    -0.032385408076210126,
    -17.20744048260943,
    60.223125,
    -1.3287857969718009,
    0.34356041660314957,
    9.0501517903694
])

# =========================
# WEEK 2 OUTPUTS
# =========================
week2_outputs = np.array([
    1.1867858443665185e-41,
    0.2715258567299176,
    -0.1198581070659559,
    -13.082213230390916,
    50.179390256321376,
    -1.5080002951000964,
    0.31639485635485903,
    9.0219949134074
])

def svm_inspired_week3(w1, w2, y1, y2, alpha=0.35):
    """
    SVM-inspired margin rule:
    - Choose the better of Week 1 and Week 2 as the positive side
    - Move slightly farther away from the worse point in that direction
    - Clip to [0,1]

    This is not a true trained SVM, but it mimics the idea of moving
    deeper into the high-performance side of a separating margin.
    """

    # both essentially zero -> not enough signal, use midpoint + mild exploration
    if abs(y1) < 1e-20 and abs(y2) < 1e-20:
        midpoint = (w1 + w2) / 2
        perturb = np.zeros_like(w1)
        perturb[0] = 0.15
        if len(w1) > 1:
            perturb[1] = -0.15
        candidate = midpoint + perturb
        return np.clip(candidate, 0, 1)

    # identify better and worse point
    if y2 > y1:
        best = w2
        worse = w1
    else:
        best = w1
        worse = w2

    # move beyond the better point in the same direction away from worse point
    direction = best - worse
    candidate = best + alpha * direction

    return np.clip(candidate, 0, 1)

# Generate Week 3
week3_inputs = []
for i in range(8):
    w3 = svm_inspired_week3(
        week1_inputs[i],
        week2_inputs[i],
        week1_outputs[i],
        week2_outputs[i],
        alpha=0.35
    )
    week3_inputs.append(np.round(w3, 6))

# Print results
print("Suggested Week 3 inputs (arrays):")
for i, arr in enumerate(week3_inputs, start=1):
    print(f"Function {i}: {arr}")

print("\nPortal-ready format:")
for i, arr in enumerate(week3_inputs, start=1):
    portal_string = "-".join(f"{x:.6f}" for x in arr)
    print(f"Function {i}: {portal_string}")
```
